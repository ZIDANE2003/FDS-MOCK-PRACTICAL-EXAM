```
 # MOCK PRACTICAL EXAM (NO 5):
m= int(input("enter number of rows: "))
n = int(input("enter number of columns: "))
Firstmat = [[0 for col in range(n)] for row in range(m)]
for i in range(m):
    for j in range(n):
        num = int(input("Enter  elements in first matrix: "))
        Firstmat[i][j]= num

print("The first matrix is:")
print(Firstmat)

Secondmat = [[0 for col in range(n)] for row in range(m)]
for i in range(m):
    for j in range(n):
        num = int(input("Enter  elements in second matrix: "))
        Secondmat[i][j]= num

print("The second matrix is:")
print(Secondmat)
# addition of two matrices
def addaddition_matrix(Firstmat,Secondmat):
    result = [[Firstmat[i][j] + Secondmat[i][j]  for j in range(len(Firstmat[0]))] for i in range(len(Firstmat))]

    print("The Addition of Two Matrices:")
    for r in result:  
        print(r)
 
#subtraction of two matrices
def subtraction_matrix(A,B):
    result = [[Firstmat[i][j] - B[i][j]  for j in range(len(Firstmat[0]))] for i in range(len(Firstmat))]

    print("The Subtraction of Two Matrices:")
    for r in result:
        print(r)

#Multiplication of two matrices
def Multiplication_matrix(Firstmat,Secondmat):
    result = [[0 for col in range(n)] for row in range(m)]
    for i in range(len(Firstmat)):  
       for j in range(len(Secondmat[0])):  
           for k in range(len(Secondmat)):  
               result[i][j] += Firstmat[i][k] * Secondmat[k][j]  

    print("Matrix Multiplication is: ")
    for r in result:  
        print(r) 
#Transpose 
def transpose():
    result = [[0 for col in range(n)] for row in range(m)] 
    for i in range(len(Firstmat)):  
       for j in range(len(Firstmat[0])): 
           result[j][i] = Firstmat[i][j]  
    print("Transpose of First Matrix is:")
    for r in result:  
        print(r)  
    for i in range(len(Secondmat)):  
       for j in range(len(Secondmat[0])):  
           result[j][i] = Secondmat[i][j]  
  
    print("Transpose of Second  Matrix is:")
    for r in result:  
        print(r)  

addaddition_matrix(Firstmat,Secondmat)  
subtraction_matrix(Firstmat,Secondmat)
Multiplication_matrix(Firstmat,Secondmat)
transpose()
```
